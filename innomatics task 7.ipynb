```

class SmartLight:
    def __init__(self, name):
        self.name = name
        self.state = "OFF"
    def turn_on(self):
        self.state = "ON"
    def turn_off(self):
        self.state = "OFF"
    def status(self):
        print(f"{self.name} is {self.state}")

light = SmartLight("Bedroom Light")
light.turn_on()
light.status()
```

```

class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department
    def display(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")

emp = Employee("Rahul", "EMP102", "IT")
emp.display()
```

```

class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone
    def display(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")

c = Contact("Anita", "9876543210")
c.display()
```

```

class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price
    def display(self):
        print(f"Product: {self.name}")
        print(f"Price: ₹{self.price}")

p = Product("Headphones", 2499)
p.display()
```

```

class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating
    def display(self):
        print(f"Movie: {self.name}")
        print(f"Rating: {self.rating} / 5")

m = Movie("Inception", 4.8)
m.display()
```

```

class Delivery:
    def __init__(self, customer, address):
        self.customer = customer
        self.address = address
    def display(self):
        print("Delivery Details")
        print(f"Customer: {self.customer}")
        print(f"Address: {self.address}")

d = Delivery("Suman", "Hyderabad")
d.display()
```
