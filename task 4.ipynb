## Problem 1: Mobile Recharge Validation System

```

def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        print("Recharge amount must be at least ₹50")
        return False
    elif amount not in valid_plans:
        print("Invalid plan. Available plans:", valid_plans)
        return False
    else:
        print("Recharge successful for ₹", amount)
        return True


while True:
    amt = int(input("Enter recharge amount: "))
    if validate_recharge(amt):
        break
    else:
        print("Please try again...")
```

## Problem 2: Inventory Reorder Alert System

```

def check_inventory(stock):
    for product, quantity in stock.items():
        if quantity < 15:
            print(product, ": Reorder Alert")
        else:
            print(product, ": Stock OK")


inventory = {
    "Laptop": 10,
    "Mouse": 25,
    "Keyboard": 12,
    "Monitor": 20
}

check_inventory(inventory)
```

## Problem 3: Student Result Processing System

```

def student_result(marks):
    total = 0
    for m in marks:
        total += m

    average = total / len(marks)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"


marks_list = [60, 55, 40, 70, 45]
print("Result:", student_result(marks_list))
```

## Problem 4: Cab Fare Estimator with Retry Option

```

def calculate_fare(distance, peak):
    fare = 50 + (distance * 12)

    if peak == "yes":
        fare += fare * 0.25

    return fare


while True:
    km = float(input("Enter distance (km): "))
    peak_hour = input("Is it peak hour? (yes/no): ").lower()

    total_fare = calculate_fare(km, peak_hour)
    print("Total Fare: ₹", total_fare)

    retry = input("Do you want to calculate again? (yes/no): ").lower()
    if retry != "yes":
        break
```

## Problem 5: Employee Attendance Eligibility Checker

```

def attendance_check(attendance):
    present = 0

    for day in attendance:
        if day == "P":
            present += 1

    percentage = (present / len(attendance)) * 100

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


attendance_list = ["P", "P", "A", "P", "P", "A", "P"]
print("Attendance Status:", attendance_check(attendance_list))
```

## Problem 6: Password Strength Checker

```

def password_strength(password):
    if len(password) < 8:
        return "Weak Password"

    has_digit = False
    has_special = False

    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in "@#$":
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"


pwd = input("Enter password: ")
print(password_strength(pwd))
```
