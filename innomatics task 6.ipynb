```

def parking_lot(capacity, logs):
    count = 0
    peak = 0
    for log in logs:
        if log == "IN":
            count += 1
        elif log == "OUT" and count > 0:
            count -= 1
        peak = max(peak, count)
    status = "Full" if count >= capacity else "Available"
    print("Currently Parked Vehicles:", count)
    print("Parking Status:", status)

parking_lot(50, ["IN","IN","IN","OUT","IN","IN","OUT"])
```

```

def delivery_time(distance, traffic, weather):
    time = distance * 5
    if traffic == "High":
        time += 15
    elif traffic == "Medium":
        time += 8
    if weather == "Rainy":
        time += 10
    elif weather == "Stormy":
        time += 20
    print("Estimated Delivery Time:", time, "minutes")

delivery_time(8, "High", "Rainy")
```

```

def seat_occupancy(total, booked):
    occupied = len(booked)
    percent = (occupied / total) * 100
    status = "Housefull" if percent == 100 else "Almost Full" if percent >= 75 else "Available"
    print("Occupancy:", int(percent), "%")
    print("Show Status:", status)

seat_occupancy(200, [1]*150)
```

```

def server_load(readings):
    avg = sum(readings) / len(readings)
    if avg < 50:
        status = "Normal"
    elif avg <= 80:
        status = "Warning"
    else:
        status = "Critical"
    print("Average CPU Load:", int(avg), "%")
    print("Server Status:", status)

server_load([45,60,70,85,90])
```

```

def resource_monitor(usage):
    overused = []
    for r,h in usage.items():
        if h > 8:
            overused.append(r)
    print("Overused Resources:", ", ".join(overused))
    print("Energy Alert:", "Yes" if overused else "No")

resource_monitor({"Projector":6,"AC":9,"Lights":4})
```

```

def event_registration(capacity, registrations):
    confirmed = min(capacity, registrations)
    waitlist = max(0, registrations - capacity)
    status = "Closed" if registrations >= capacity else "Open"
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)
    print("Registration Status:", status)

event_registration(100,105)
```
